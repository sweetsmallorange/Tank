```python
import pandas as pd
from pandas.errors import EmptyDataError
from pathlib import Path
```

```python

basins_list = pd.read_csv("/data2/zw/sac/data/pet_basins.txt", dtype=str, header=None)

basins_list = basins_list[0].values.tolist()
```

```python
output_root = Path("/home/zhuwu/Dataset/CAMELS") / "model_output_daymet" / "model_output" / "flow_timeseries" / "daymet"
dfs = []
need = ["PRCP", "PET", "ET", "tmax", "tmin", "vp", "OBS_RUN"]

for idx, basin in enumerate(basins_list):
    print(basin)
    files = list(output_root.glob(f"**/{basin}_*_model_output.txt"))
    if len(files) == 0:
        raise RuntimeError(f"No discharge file found for Basin {basin}")
    df_total = pd.read_csv(files[0], sep=r"\s+")
    dates = df_total.YR.map(str) + "/" + df_total.MNTH.map(str) + "/" + df_total.DY.map(str)
    df_total.index = pd.to_datetime(dates, format="%Y/%m/%d")
    # assert len(need) == 3

    for i in range(1, len(files)):
        try:
            df = pd.read_csv(files[i], sep=r"\s+")
            dates = df.YR.map(str) + "/" + df.MNTH.map(str) + "/" + df.DY.map(str)
            df.index = pd.to_datetime(dates, format="%Y/%m/%d")
            # assert len(need) == 3
            df_total = df_total + df
        except EmptyDataError:
            print(f"No columns to parse from file {files[i]}")

    df_total['PET'] = df_total['PET'] / len(files)
    df_total['pet'] = df_total['PET']
    df_total = df_total["1980-01-01":"2008-12-31"]

    file = Path("/data2/zw/dataset/Uniform-CAMELS/CAMELS-US/forcing") / f"{basin}.csv"
    df_origin = pd.read_csv(file, sep=",", header=0)
    df_origin["date"] = pd.to_datetime(df_origin["date"], format="%Y-%m-%d")
    df_origin = df_origin.set_index("date")

    df_final = pd.concat((df_total, df_origin), axis=1)
    df_final = df_final[["prcp", "pet", "srad", "tmax", "tmin", "vprp"]]
    df_final.to_csv(Path("/data2/zw/dataset/Uniform-CAMELS/CAMELS-US_pet/forcing") / f"{basin}.csv",
                    index_label="date", float_format="%.2f")
# print(df_total)
# df_final
```
